Data cLeaning and Tranformations         ( Input missing values, encoding categorical variables etccc )

Encoding Categorical Data ( adult male )

```python
print(df['adult_male'].value_counts())
```

```python
# as adult male col contains
# it has two unique values true and false
# which can be converted to numberical data which helps ML models to understand the data better
# df['cloumn_name'].replace({oldvalue:newvalue}, inplace = True)
df['adult_male'].replace({True:1,False:0}, inplace = True)
```

Missing Values

```python
 print(df.isnull().sum())
```


First Appoach : Deletion

```python
# But deletion can cause losing important data if its not MCAR
# df.dropna(subset = ['Age'], inplace = True)
```

Second Apporach : Imputation

```python
# this method is good but
# We should check for any outliers because Some central tendency are sensitive to outliers
# MEAN, MEDIAN, MODE
```

```python
# to identify the outliers you can use box plot
sns.boxplot(y = 'age',data = df)
plt.title("Box Plot for Age")
plt.show()
```

Median Imputation : AGE

```python
# as in the above boxplot we can see that age col has outliers
# we are gng to use median imputation
df['age'] = df['age'].fillna(df['age'].median())
# here fillna is used to fill missing values
# here df['age'].median() calculates median of age col
```

```python
mmmm = df['deck'].mode()
print(mmmm)
```

Deck col : Categorical data : Mode Imputation

```python
# as deck is categorical data we are using mode imputation
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])
```

Embark_town : Categorical data

```python
# as it has low missing values we can drop the rows as it has minimum effect
df.dropna(subset = ['embark_town'],inplace = True)
```

```python
 print(df.isnull().sum())
```

Changing Col Name

```python
df.rename(columns = {'deck':'cabin'},inplace = True)
```

Histogram for who columns

```python
#histogram is graphical representation of distribution
plt.hist(df['who'],color = 'skyblue', edgecolor = 'black')
plt.xlabel("Passengers type")
plt.ylabel("Frequency")
plt.title('Distribution of passengers type')
plt.show()
```

Co-relation Matrix

```python

num_df = df.select_dtypes(include = ['int64','float64'])
corr_mat = num_df.corr()
plt.figure(figsize = (10,8))
sns.heatmap(corr_mat, annot = True, cmap = 'PuBuGn',fmt = ".2f")
# here annot adds numerical annotation displaying the numerical values
#here cmap indictes the color map used in the heatmap
# gn(green)->high :( 1 or -1) & Bn(blue)-> moderate (0.3 to 0.7) & Pu(purple)->low (0.1 to 0.3) correlations
# here fmt = ".2f" indicates that decimal values upto two places
plt.title("Correlation Matrix of Titanic Dataset")
plt.show()
```
